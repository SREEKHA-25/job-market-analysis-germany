```python
import zipfile
import pandas as pd
from pathlib import Path
```

```python
zip_path = Path("C:/Users/sivanisree/Downloads/Job Posting.csv.zip")
extract_path = Path("data/extracted")  # Where we’ll extract the file

# Extract the ZIP
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("ZIP extracted!")
```

```python
csv_file = Path("data/extracted/Job Posting.csv")

# Load the CSV file into a DataFrame
df = pd.read_csv(csv_file, encoding='ISO-8859-1')

# Show basic info
print("Rows:", df.shape[0], "| Columns:", df.shape[1])
df.head(9919)
```

```python
european_countries = [
    "Germany", "France", "Netherlands", "Belgium", "Austria",
    "Switzerland", "Sweden", "Denmark", "Poland", "Italy", "Spain"
]

# Step 2: Filter rows where 'Location' contains any of the countries
df_europe = df[df['Location'].str.contains('|'.join(european_countries), case=False, na=False)]

# Step 3: Show shape of filtered data
print("Filtered European job listings:", df_europe.shape[0])
df_europe.head(9918)
```

```python
df_europe['Location'].value_counts().head(1958)
```

```python
df_europe = df_europe.copy()  
df_europe['Country'] = df_europe['Location'].apply(lambda x: x.split(",")[-1].strip())
```

```python
df_europe['Country'].unique()
```

```python
df_europe['Country'].value_counts()
```

```python
import matplotlib.pyplot as plt

# Count the number of job listings per country
country_counts = df_europe['Country'].value_counts()

# Create horizontal bar chart
plt.figure(figsize=(6,4))
country_counts.plot(kind='barh', color='skyblue')

# Add titles and labels
plt.title('European Job Listings by Country')
plt.xlabel('Number of Job Listings')
plt.ylabel('Country')

# Make the highest country appear at the top
plt.gca().invert_yaxis()

# Improve layout
plt.tight_layout()

# Show the chart
plt.show()
```

```python
df_europe = df_europe.rename(columns={"Job Opening Title": "Job_Title"})
```

```python
df_europe['Job_Title'].head(10)
```

```python
df_europe['Job_Title'].value_counts().head(15)
```

```python
import re

# Remove gender labels like (m/f/d), (w/m/x), etc.
df_europe['Job_Title_Clean'] = df_europe['Job_Title'].apply(
    lambda x: re.sub(r'\(.*?\)', '', str(x)).strip()
)

# View cleaned titles
df_europe['Job_Title_Clean'].value_counts().head(15)
```

```python
import warnings
warnings.filterwarnings("ignore", category=UserWarning)
import matplotlib.pyplot as plt
Job_counts = df_europe['Job_Title'].value_counts().head(15)
plt.figure(figsize=(6,4))
Job_counts.plot(kind='barh', color='salmon')         
plt.title(' Top 15 Job Titles in European Listings')
plt.xlabel('Number of Job Listing')
plt.ylabel('Job Title')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
top_companies = df_europe['Website Domain'].value_counts().head(15)

# Plot the top 15
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
top_companies.plot(kind='barh', color='mediumseagreen')

plt.title('Top 15 Companies Hiring in Europe')
plt.xlabel('Number of Job Listings')
plt.ylabel('Company Website')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
skills = ['python', 'sql', 'excel', 'power bi', 'tableau', 'remote']

# Convert all descriptions to lowercase for easy matching
df_europe['Description'] = df_europe['Description'].fillna('').str.lower()

# Count how many job descriptions mention each skill
skill_counts = {skill: df_europe['Description'].str.contains(skill).sum() for skill in skills}

# Show as DataFrame
import pandas as pd
skill_df = pd.DataFrame.from_dict(skill_counts, orient='index', columns=['Job Count'])
skill_df.sort_values(by='Job Count', ascending=True).plot(kind='barh', figsize=(6, 4), color='orchid')

plt.title("Demand for Skills in European Job Listings")
plt.xlabel("Number of Job Listings")
plt.tight_layout()
plt.show()
```
